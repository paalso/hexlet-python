```python
train_data.index = train_data.PassengerId
train_data.drop(['PassengerId'], axis=1, inplace=True)
print(train_data.index)
print(train_data.index.name)
```

```python
# Была ли у пассажира своя каюта?
train_data['Has_Cabin'] = train_data["Cabin"].apply(lambda x: 0 if type(x) == float else 1)
train_data.head()
```

```python
# Какого размера была семья пассажира?
train_data['FamilySize'] = train_data['SibSp'] + train_data['Parch'] + 1
train_data.head()
```

```python
# Пассажир был на Титанике один или с попутчиками?
train_data['Is_alone'] = train_data["FamilySize"].apply(lambda x: 0 if x > 1 else 1)
train_data.head()
```

```python
# Используем регулярные выражения, чтобы найти сокращения Ms, Mr или другие обращения
def get_title(name):
    title_search = re.search(' ([A-Za-z]+)\.', name)
    if title_search:
        return title_search.group(1)
    return ""
```

```python
# Применяем к столбцу Name
train_data['Title'] = train_data['Name'].apply(get_title)
train_data.head()
```

```python
train_data['Title'].unique()
```

```python
# Исправляем опечатки и группируем редкие обращения
train_data['Title'] = train_data['Title'].replace(
    ['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'],
    'Rare'
)
train_data['Title'] = train_data['Title'].replace('Mlle', 'Miss')
train_data['Title'] = train_data['Title'].replace('Ms', 'Miss')
train_data['Title'] = train_data['Title'].replace('Mme', 'Mrs')
train_data.head()
```

```python
train_data['Title'].unique()
```

```python
# Очищаем данные от обработанных столбцов
train_data = train_data.drop(['Name','Ticket','Cabin'], axis = 1)
train_data
```

```python
def count_by_group(df, group_column, share=False):
    """
    Считает количество элементов в каждой группе для указанного столбца.

    Parameters:
    df (pd.DataFrame): Датафрейм с данными.
    group_column (str): Название столбца для группировки.
    share (bool): Если True, возвращает долю каждой группы относительно общего числа элементов.

    Returns:
    pd.DataFrame: Датафрейм с количеством элементов в каждой группе и, при необходимости, долей.
    """
    # Группировка и подсчет количества элементов в каждой группе
    counted_by_group = df.groupby(group_column).size().reset_index(name='Count')
    
    # Установка столбца группировки как индекса
    counted_by_group.set_index(group_column, inplace=True)

    # Если нужно, добавляем долю
    if share:
        total_number = df[group_column].count()
        counted_by_group['Share'] = counted_by_group['Count'] / total_number

    return counted_by_group
```

```python
import matplotlib.pyplot as plt

x = list(range(10))
x = [-i for i in x[::-1]] + x

y_1 = [i for i in x]
y_2 = [i**2 for i in x]
y_3 = [i**3 for i in x]
```

```python
def get_polynomial_subplots():
    fig, axes = plt.subplots(1, 3)
    fig.set_size_inches(12, 4)
    fig.suptitle('Polynomial')

    axes[0].plot(x, y_1, color="b", label='linear')
    axes[0].legend(loc='upper right')
    axes[0].set_xlabel('x-values')
    axes[0].set_ylabel('y-values')
    axes[0].grid(linestyle=':')  # Точечная сетка


    axes[1].plot(x, y_2, color="b", label='square')
    axes[1].legend(loc='upper right')
    axes[1].set_xlabel('x-values')
    axes[1].set_ylabel('y-values')
    axes[1].grid(linestyle=':')  # Точечная сетка


    axes[2].plot(x, y_3, color="b", label='cubic')
    axes[2].legend(loc='upper right')
    axes[2].set_xlabel('x-values')
    axes[2].set_ylabel('y-values')
    axes[2].grid(linestyle=':')  # Точечная сетка
    
    plt.show()
```

```python
from dash.dependencies import Input, Output

app = dash.Dash(__name__)

app.layout = html.Div(children=[
    html.Button('Change color', id='button'),
])

@app.callback(
    Output('button', 'style'),
    [Input('button', 'n_clicks')]
)
def update_color(n_clicks):
    if n_clicks is None:
        n_clicks = 0
    if n_clicks % 2 == 0:
        return {'background-color': 'red'}
    else:
        return {'background-color': 'blue'}

if __name__ == '__main__':
    app.run_server(debug=True)
```
